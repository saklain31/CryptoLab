```python
def multi(a1,a2):
    a1 = a1.zfill(8)
    ln = len(a2)
    st = "0".zfill(8)
    
    for i in range(ln):
        if(a2[ln-i-1]=="1"):
            temp = mulShift(a1,i)
        else:
            temp = mulShift("0".zfill(len(a1)),i)
        
        st = bxor(st.zfill(max(len(st),len(temp))),temp.zfill(max(len(st),len(temp))))
        
    return st
        
        
def bitPos(div):
    div0 = -1
    for i in range(len(div)):
        if div[i]=="1":
            div0=i
            break
    return div0

def mod(div,val):
    val = val.zfill(max(len(div),len(val))
    div0 = bitPos(div)
    val0 = bitPos(val)
    
    sft = val0 - div0 #NEG
    #if sft<0:
     #   return div
    
    
    #print("sft",sft)
    tempval = binShift(val,sft)
    #print("***")
    div = bxor(div,tempval)
    
    
    #print(div,val)
    
    while int(div,2)>=int(val,2):
        sft = val0 - div0 #NEG
        tempval = binShift(val,sft)
        div = bxor(div,tempval)
        
        div0 = bitPos(div)
        val0 = bitPos(val)
          
    #print(sft)
    return div
```

```python
def imult(x,y):
    binary = bin(y)[2:]
    i = 0
    sum = 0
    index = len(binary)-1
    while index >= 0:
        #print("index : "+str(index))
        if binary[index] == '1':
            sum = sum ^ (x << i)
        i = i + 1
        index = index - 1
    #print("sum = "+ str(sum))
    return sum

def imod(x,y):
    while len(bin(x)) >= len(bin(y)):
        lack = len(bin(x))-len(bin(y))
        x = x ^ (y << lack)
    return x



multi("11010111","10001101")
hexlist = [hex(x)[2:] for x in range(256)]

temphex = []
temp = []
for i in range(256):
    temp.append(hexlist[i])
    if i%16==15:
        temphex.append(temp)
        temp = []
    
#for i in range(16):
 #   print(temphex[i])
```
