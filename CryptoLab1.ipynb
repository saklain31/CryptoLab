```python
#Ceaser Cipher

file = open("/home/saklain/CryptoLab/ceaser.txt","r")
fileout = open("/home/saklain/CryptoLab/ceaout.txt","w")

for line in file:
    st=""
    for i in line:
        if(i==' '):
            continue
        if(i=='\n'):
            print("newline")
            continue
        #print(i)
        charid = (ord(i)+3)%123
        if(charid<97):
            charid+=97
        
        st= st+chr(charid)
    
    #print(st)
    fileout.write(st+"\n")
    
    
file.close()

fileout.close()                   
    
```

```python
#ENCRYPTION
fileout = open("/home/saklain/CryptoLab/cr_out.txt","w")

key="pzRruAiO"
idx=0

with open("/home/saklain/CryptoLab/ceaser.txt","r") as f:
    while True:
        ch = f.read(1)
        if not ch:
              print("End of file")
              break
        if ch=='\n':
            break
        if ch==' ':
            fileout.write(str(' '))
            continue

        plainidx = ord(ch)
        if(plainidx>90):
            plainidx-=6
        plainidx-=65
        
        print(key[idx])
        keyidx = ord(key[idx])
        if(keyidx>90):
            keyidx-=6
        keyidx-=65
        
        idx+=1
        if(idx==len(key)):
            idx=0
            
        #print(ch,ord(ch),plainidx)
        newidx = (plainidx+keyidx)%52
                                
        cipidx = newidx+65
        
        if(cipidx>90):
            cipidx+=6
        
        print(ch,ord(ch),plainidx,cipidx,chr(cipidx))
        
        
        fileout.write(str(chr(cipidx)))
        
fileout.close()
```
