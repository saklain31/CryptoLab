```python
def bxor(a1,a2):
    ln=len(a1)
    a1 = bin(int(a1, 16))[2:].zfill(ln*4)
    a2 = bin(int(a2, 16))[2:].zfill(ln*4)
    y = int(a1,2) ^ int(a2,2)
    return hex(int(('{0:b}'.format(y)).zfill(ln*4), 2))[2:].zfill(ln)


def isReadable(st): 
    return bool(re.search('^[a-zA-Z0-9\., \'\"\-_\:\(\)]+$', st))

def isValid(st):
    if str("\\x") in str(st):
        return False
    else:
        return True
```
